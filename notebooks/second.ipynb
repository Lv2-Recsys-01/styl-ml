```python
def removebubble(row):
    bubble={'컬러의','컬러','무신사','룩','매칭한','매칭','와'}
    new_tags=set(list(row['gpttag']).copy())
    return list(new_tags-bubble)

def tolist_withspace(row):
    return list(row['gpttag'].rstrip("]").lstrip("[").replace(" '","").replace(' "','').replace('"','').replace("'","").replace("컬러의 ",'').replace("컬러 ",'').split(","))
#return list(row['gpttag'].rstrip("]").lstrip("[").replace(' ','').replace('"','').replace("'","").replace("컬러의",'').replace("컬러",'').split(","))
def tolist(row):
    return list(row['gpttag'].rstrip("]").lstrip("[").replace('"','').replace("'","").replace("컬러의",'').replace("컬러",'').split(","))
```

```python
df['gpttag']=df.apply(tolist_withspace,axis=1)
df['gpttag']=df.apply(removebubble,axis=1)
```
